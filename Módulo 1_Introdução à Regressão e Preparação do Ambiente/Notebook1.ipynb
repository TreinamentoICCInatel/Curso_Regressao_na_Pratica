### Importando bibliotecas
Importar uma biblioteca significa tornar suas funções e ferramentas disponíveis para uso no seu código.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

Vamos fazer um pequeno teste adicionando um arquivo na pasta e importando com o pandas

```python
carros = pd.read_csv('carros.csv')
```

Agora o dataset está disponivel para uso, vamos usar um comando para fazer uma primeira análise do dataset



```python
print(carros.head())
```

Esse comando exibiu as primeiras linhas do dataset.
Podemos observar que ele contém a quilometragem e o preço de carros.
Agora, vamos plotar um gráfico simples para visualizar a relação entre essas variáveis.

```python
sns.scatterplot(x="quilometragem", y="preco", data=carros)
plt.title("Relação entre quilometragem e preço de carros")
plt.xlabel("Quilometragem (km)")
plt.ylabel("Preço (R$)")
plt.show()
```

O gráfico mostra uma relação inversa entre quilometragem e preço:
conforme a quilometragem aumenta, o preço do carro tende a diminuir.
Essa é a base para aplicarmos um modelo de regressão linear — e representa nossa primeira análise exploratória de dados!
